```python
## Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
```

```python
## Load csv file
link_covid_data = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'

data = pd.read_csv(link_covid_data)
```

```python
## Definisikan negara-negara
indo=data.query('iso_code == "IDN"')
spain=data.query('iso_code == "ESP"')
german=data.query('iso_code == "DEU"')
argentina=data.query('iso_code == "ARG"')
```

```python
## gunakan plot untuk menampilkan data
sns.lineplot(data=indo,x='date',y='total_deaths')
sns.lineplot(data=spain,x='date',y='total_deaths')
sns.lineplot(data=german,x='date',y='total_deaths')
sns.lineplot(data=argentina,x='date',y='total_deaths')
plt.xticks([10,120,230,340,450,562], rotation=45)
plt.legend(['Indonesia', 'Spanyol', 'German', 'Argentina'])
plt.show()
```

```python
## Definisikan Negara mulai dari tanggal 1 mei sampai hari ini
indo2      = indo.query('date >= "2021-05-01"')
spain2     = spain.query('date >= "2021-05-01"')
german2    = german.query('date >= "2021-05-01"')
argentina2 = argentina.query('date >= "2021-05-01"')
```

```python
## gunakan plot untuk menampilkan data
sns.lineplot(data=indo2,x='date',y='total_deaths')
sns.lineplot(data=spain2,x='date',y='total_deaths')
sns.lineplot(data=german2,x='date',y='total_deaths')
sns.lineplot(data=argentina2,x='date',y='total_deaths')
plt.xticks([7,37,67,97,127,158], rotation=45)
plt.legend(['Indonesia', 'Spanyol', 'German', 'Argentina'])
plt.show()
```
